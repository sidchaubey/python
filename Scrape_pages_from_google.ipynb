```
import requests
from bs4 import BeautifulSoup as bs

# text = input("Enter the keyword to be scraped: ")# Please Enter racial discrimination covid19
page1 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk032vRCMhFwWBkrjiR3kAVu15wq6FQ:1590911387398&ei=m2HTXuv3F8ve9QOl_piIAQ&start=0&sa=N&ved=2ahUKEwirsfXgzt3pAhVLb30KHSU_BhE4ChDy0wN6BAgLEC8&biw=639&bih=600"
page2 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk03nRzVYFsJVMDnN2aFPrQy56Hu85A:1590914218371&ei=qmzTXsegFpTbrQHnkbvwDg&start=10&sa=N&ved=2ahUKEwjHneqm2d3pAhWUbSsKHefIDu4Q8tMDegQICxAv&biw=639&bih=600"
page3 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk02CwICJMzHZjIBMbNzsYKUB5XaIeg:1590914247120&ei=x2zTXujuBo2o9QPbk5rYBw&start=20&sa=N&ved=2ahUKEwio7sS02d3pAhUNVH0KHduJBns4ChDy0wN6BAgLEDI&biw=639&bih=600"
page4 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk03PrzUxLCu7Uo79U2SZekd85swtlA:1590914391857&ei=V23TXtfxM4WR9QOIxr-oDA&start=30&sa=N&ved=2ahUKEwiX-cb52d3pAhWFSH0KHQjjD8U4FBDy0wN6BAgLEDE&biw=639&bih=600"
page5 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk02c-u6ENE9_HoxYO9nvBS0JKqN7Nw:1590914446463&ei=jm3TXpLfG6fgz7sPveqRoAo&start=40&sa=N&ved=2ahUKEwiS3suT2t3pAhUn8HMBHT11BKQ4HhDy0wN6BAgLEDM&biw=639&bih=600"
page6 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk03KeMkDA8JuFyFGG1KcxGar8eUqEg:1590914479396&ei=r23TXqfdF_OU4-EP8YyXkAU&start=50&sa=N&ved=2ahUKEwjn8KWj2t3pAhVzyjgGHXHGBVI4KBDy0wN6BAgLEDU&biw=639&bih=600"
page7 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk003LOZ3AP-ADcCtvKMsCGPnb026Pg:1590914508084&ei=zG3TXsDSBPnCz7sPpsyRiAs&start=60&sa=N&ved=2ahUKEwjA6Pyw2t3pAhV54XMBHSZmBLE4MhDy0wN6BAgLEDc&biw=639&bih=600"
page8 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk01JY7Zd6avZthpDOHPEGXgcm9ZZwg:1590914533746&ei=5W3TXruXLfXD3LUPjMepuAk&start=70&sa=N&ved=2ahUKEwj7nZu92t3pAhX1IbcAHYxjCpc4PBDy0wN6BAgLEDc&biw=639&bih=600"
page9 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk03fIrMJsK_Tj9OMgzEMMjyRYg9QAg:1590914555438&ei=-23TXpSxGvSzmge277bwDg&start=80&sa=N&ved=2ahUKEwjUmsfH2t3pAhX0meYKHba3De44RhDy0wN6BAgLEDc&biw=639&bih=600"
page10 = "https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&sxsrf=ALeKk01FLDt8UJkgSyW-QSZajhCXGeuniw:1590914582778&ei=Fm7TXpL7LseY4-EP55SFyAQ&start=90&sa=N&ved=2ahUKEwiS3svU2t3pAhVHzDgGHWdKAUk4UBDy0wN6BAgLEDc&biw=639&bih=600"


pages = [page1,page2,page3,page4,page5,page6,page7,page8,page9,page10] # List of Pages
for i in range(len(pages)):# Iterating through list of pages
  r = requests.get(pages[i])# Requesting to get i url
  html = r.content# getting content
  soup = bs(html,"html")# using beautifulsoup to parse through html
  if i==0:# This will initilize links to empty list only when we're passing through first page
    links = []
  links.append(soup.find_all('div',{'class':"kCrYT"}))# href link is associated with div class "kCrYT"
  post = soup.find_all("div",{"class":"BNeawe vvjwJb AP7Wnd"})# Heading is associated with div and class name is BNeawe vvjwJb AP7Wnd
  if i==0:# We only want to initialize heading to empty list for first iteration
    heading = []
  for j in post:# iterating through each post element
    heading.append(re.sub('<[^>]*>', ' ', str(j)))# removing tag associated with each article
  
  post = soup.find_all("div",{"class":"BNeawe s3v9rd AP7Wnd"})# Metadata content is associated in div and class name is "BNeawe s3v9rd AP7Wnd"
  if i==0:# We only want to initialize content to empty list for first iteration
    content=[]
  for j in post:# Iterating through each post 
    content.append(j.text)# Appending only the text associated with each article
    
  
   

```

```
df2 = df2.drop_duplicates(subset="content",keep="first",ignore_index=True)# Removing the duplicate entry
df2
```

```
lst_links = []
for i in links:# Iterating through each list of 10 page link
  for j in i:# Iterating through each article
    lst_links.append(re.findall("href=[\"\'](.*?)[\"\']", str(j)))# Only selecting the href part of each article
```
