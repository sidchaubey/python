```
import requests
from bs4 import BeautifulSoup as bs

# text = input("Enter the keyword to be scraped: ")# Please Enter racial discrimination covid19
r = requests.get("https://www.google.com/search?q=anti+racial+case+covid19&rlz=1C1GCEB_enIN889IN889&oq=anti+racial+case+covid19&aqs=chrome..69i57.11076j0j1&sourceid=chrome&ie=UTF-8")
```

```
html =r.content
```

```
soup = bs(html,"html")
```

```
print(soup.prettify())
```

```
soup.find_all('div',{'class':"kCrYT"})
```

```
post = soup.find_all("div",{"class":"BNeawe vvjwJb AP7Wnd"})
post
```

```
q = []
for i in post:
  q.append(re.sub('<[^>]*>', ' ', str(i)))
print(q)
```

```
df1 = pd.DataFrame(q,columns=["Heading"])
```

```
post = soup.find_all("div",{"class":"BNeawe s3v9rd AP7Wnd"})
post
```

```
q = []
for i in post:
  q.append(i.text)
print(q)
```

```
df2 = pd.DataFrame(q,columns=["content"])
df2
```

```
df2 = df2.drop_duplicates(subset="content",keep="first",ignore_index=True)
```

```
df1
```

```
# Code review required for link 

links_with_text = []
for a in soup.find_all('a', href=True): 
    if a.text: 
        links_with_text.append(a['href'])
links_with_text
```

```
df3 = pd.concat([df1,df2],axis=1)
df3
```

```
df3.to_csv('AntiRacialCaseCovid19.csv', sep=',', encoding='utf-8', index=False)
```
